Qs1.What is a random variable in probability theory?
- A random variable is a numerical value that represents the outcome of a random experiment. It can be **discrete** (specific values) or **continuous** (any value in a range).
Qs2.What are the types of random variables?
- The two main types of random variables are:

1. **Discrete random variable** – takes specific, separate values (e.g., number of heads in coin tosses).
2. **Continuous random variable** – takes any value within a range (e.g., height, temperature).
Qs3.What is the difference between discrete and continuous distributions/
- **Discrete distributions** describe probabilities of specific, separate values (e.g., binomial distribution), while **continuous distributions** describe probabilities over intervals of values (e.g., normal distribution).
Qs4.What are probability distribution functions (PDF)?
- A **probability distribution function (PDF)** describes how probabilities are distributed over the values of a random variable. For discrete variables, it's called a **probability mass function (PMF)**, and for continuous variables, it's the **probability density function (PDF)**.
Qs5.How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
- A **PDF** shows the likelihood of a random variable taking a specific value (for continuous variables), while a **CDF** gives the probability that the variable is **less than or equal to** a certain value. The CDF accumulates probabilities up to that point.
Qs.6What is a discrete uniform distribution?
- A **discrete uniform distribution** is a distribution where all possible outcomes have an **equal probability** of occurring. For example, rolling a fair die (each number from 1 to 6 has a 1/6 chance).
Qs7.What are the key properties of a Bernoulli distribution/
- Key properties of a **Bernoulli distribution**:

1. **Two outcomes**: success (1) or failure (0).
2. **Single trial**.
3. **Probability of success** is $p$, and failure is $1 - p$.
4. **Mean** is $p$, **variance** is $p(1 - p)$.
Qs8.What is the binomial distribution, and how is it used in probability?
- The **binomial distribution** models the number of successes in a fixed number of **independent Bernoulli trials**, each with the same probability of success. It’s used to calculate the likelihood of getting a specific number of successes in repeated experiments, like flipping a coin multiple times.
Qs9.What is the Poisson distribution and where is it applied?
- The **Poisson distribution** models the number of times an event occurs in a fixed interval of time or space, given a constant average rate and independent occurrences. It's used in areas like call arrivals at a center or counting defects in manufacturing.
Qs.10What is a continuous uniform distribution?
- A **continuous uniform distribution** is where all values within a certain interval are **equally likely** to occur. An example is randomly picking a number between 0 and 1—each number has the same probability density.
Qs11.What are the characteristics of a normal distribution?
Here are the key characteristics of a **normal distribution**, summarized:

1. **Bell-shaped curve**: Symmetrical around the mean.
2. **Mean = Median = Mode**: All are equal and located at the center.
3. **Symmetry**: The left and right sides of the curve are mirror images.
4. **68-95-99.7 Rule**: About 68% of data falls within 1 standard deviation of the mean, 95% within 2, and 99.7% within 3.
5. **Asymptotic**: The tails approach, but never touch, the horizontal axis.
6. **Defined by two parameters**: Mean (μ) and standard deviation (σ).
Qs12.What is the standard normal distribution, and why is it important?
- The **standard normal distribution** is a special case of the normal distribution with:

* **Mean (μ) = 0**
* **Standard deviation (σ) = 1**

### Importance:

* **Simplifies calculations** using **Z-scores** (standardized values).
* **Allows comparison** between different normal distributions.
* **Used in statistical inference**, like hypothesis testing and confidence intervals.

It's a foundational tool in statistics for analyzing and interpreting data.
Qs13.What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- The **Central Limit Theorem (CLT)** states that the **sampling distribution of the sample mean** approaches a **normal distribution** as the **sample size increases**, regardless of the population's original distribution (as long as it has a finite mean and variance).

### Importance:

* **Enables use of normal distribution** for inference, even with non-normal data.
* Forms the basis for **confidence intervals** and **hypothesis testing**.
* Makes statistical analysis **more reliable and generalizable**.
Qs14.How does the Central Limit Theorem relate to the normal distribution?
- The **Central Limit Theorem (CLT)** relates to the **normal distribution** by stating that:

> As sample size increases, the **distribution of sample means** becomes approximately **normal**, even if the original population is not normally distributed.

### Key Link:

* It explains **why the normal distribution appears so often** in statistics.
* Allows use of **normal-based methods** for inference with large samples.
Qs.15.What is the application of Z statistics in hypothesis testing?
- **Z-statistics** are used in **hypothesis testing** to:

* **Compare a sample mean** to a known population mean when the population **standard deviation is known**.
* **Determine how many standard deviations** a data point or sample mean is from the population mean.

### Applications:

* **Z-test for means**
* **Z-test for proportions**
* Used to calculate the **p-value** and decide whether to **reject the null hypothesis**.
Qs16.How do you calculate a Z-score, and what does it represent?
- ### **Z-score Formula**:

$$
Z = \frac{X - \mu}{\sigma}
$$

* **X** = individual data point
* **μ** = population mean
* **σ** = population standard deviation

### **What it Represents**:

A **Z-score** tells you **how many standard deviations** a value is from the mean:

* **Z > 0**: Above the mean
* **Z < 0**: Below the mean
* **Z = 0**: Exactly at the mean

It's used to **standardize values** for comparison and probability analysis.
Qs17.What are point estimates and interval estimates in statistics?
- ### **Point Estimate**:

* A **single value** used to estimate a population parameter (e.g., sample mean to estimate population mean).

### **Interval Estimate**:

* A **range of values** (e.g., confidence interval) that likely contains the population parameter.
* Provides an **estimate with a margin of error**.

### In Short:

* **Point estimate** = precise guess
* **Interval estimate** = guess with uncertainty range
Qs18.What is the significance of confidence intervals in statistical analysis?
- ### **Significance of Confidence Intervals**:

* **Measure uncertainty** around a point estimate.
* Indicate a range where the **true population parameter likely lies**.
* Provide more **reliable and informative results** than single-point estimates.
* Commonly used to **assess precision** and **support decision-making** in research.

E.g., a 95% confidence interval means we're 95% confident the true value falls within that range.
Qs19.What is the relationship between a Z-score and a confidence interval/
- ### **Relationship Between Z-score and Confidence Interval**:

* **Z-scores** are used to **calculate confidence intervals** when the population standard deviation is known.
* Each **confidence level** corresponds to a specific **Z-score** (e.g., Z = 1.96 for 95% confidence).
* The Z-score determines how **wide the interval** is around the point estimate.

**In short**:
**Z-scores set the margin** for confidence intervals in standard normal distributions.
Qs20.How are Z-scores used to compare different distributions?
- ### **Z-scores for Comparing Distributions**:

* **Standardize values** from different distributions by converting them to a **common scale** (mean = 0, SD = 1).
* Allow **direct comparison** of scores from different units or scales.

**In short**:
Z-scores make values **comparable across different distributions** by expressing them in terms of standard deviations from their means.
Qs21.What are the assumptions for applying the Central Limit Theorem?
- The **Central Limit Theorem (CLT)** applies under these assumptions:

1. **Random Sampling**: Samples must be randomly selected.
2. **Independence**: Observations within the sample must be independent.
3. **Sample Size**: The sample size should be sufficiently large (usually n ≥ 30).
4. **Finite Variance**: The population must have a finite mean and variance.

**In short**: For the CLT to hold, you need random, independent samples with a large enough size from a population with finite variance.
Qs22.What is the concept of expected value in a probability distribution?
- ### **Expected Value**:

The **expected value** (or **mean**) of a probability distribution is the **long-term average** of all possible outcomes, weighted by their probabilities.

### Formula:

$$
E(X) = \sum (X \cdot P(X))
$$

Where:

* $X$ = possible outcomes
* $P(X)$ = probability of each outcome

### In short:

It represents the **"center"** of the distribution, or what you'd expect as the average result if the experiment were repeated many times.
Qs23.How does a probability distribution relate to the expected outcome of a random variable?
- A **probability distribution** describes all possible outcomes of a **random variable** and their associated probabilities. The **expected value** is the **weighted average** of these outcomes, where each outcome is weighted by its probability.

### In short:

The probability distribution provides the probabilities, and the expected value calculates the **average outcome** you’d expect from the distribution.
