#1.What is NumPy, and why is it widely used in Python?
- NumPy is a powerful Python library used for numerical computing. It provides support for arrays, matrices, and a large collection of mathematical functions to operate on these data structures. It's widely used because it's fast, efficient, and forms the foundation for many scientific and machine learning libraries in Python.
#2.How does broadcasting work in NumPy?
- Broadcasting in NumPy allows operations on arrays of different shapes by automatically expanding their shapes to be compatible. It avoids the need to manually reshape arrays, making computations more efficient and concise.
#3.What is a Pandas DataFrame?
- A Pandas DataFrame is a 2-dimensional, labeled data structure in Python, similar to a table in a database or an Excel spreadsheet. It allows easy data manipulation, analysis, and cleaning with labeled rows and columns.
#4.A Explain the use of the groupby() method in Pandas.
The `groupby()` method in Pandas is used to split data into groups based on some criteria, apply a function to each group (like sum, mean, etc.), and combine the results. It's useful for analyzing and summarizing data by categories.
#5.Why is Seaborn preferred for statistical visualization?
Seaborn is preferred for statistical visualizations because it simplifies complex plots, integrates well with Pandas, and provides attractive, informative visuals with minimal code—making it great for exploring data relationships and distributions.
#6.What are the differences between NumPy arrays and Python lists?
NumPy arrays are more efficient for numerical operations, support element-wise operations, and have fixed data types, while Python lists are more flexible but slower and less suitable for large-scale numerical computations.
#7.What is a heatmap, and when should it be used?
A heatmap is a data visualization that uses color to represent values in a matrix. It’s best used to show correlations, patterns, or the intensity of values across rows and columns, making it easy to spot trends or anomalies.
#8.What does the term “vectorized operation” mean in NumPy?
A vectorized operation in NumPy means performing operations on entire arrays without using explicit loops, which makes the code faster and more concise by leveraging optimized, low-level implementations.
#9.How does Matplotlib differ from Plotly?
- Matplotlib is a static, traditional plotting library ideal for basic plots and full customization, while Plotly offers interactive, web-based visualizations with built-in tools for zooming, hovering, and exporting.
#10.What is the significance of hierarchical indexing in Panda?
Hierarchical indexing in Pandas allows multiple index levels on rows or columns, enabling more complex data organization, easier data slicing, and better representation of higher-dimensional data in a 2D structure.
#11.What is the role of Seaborn’s pairplot() function?
Seaborn’s `pairplot()` function creates a matrix of scatter plots for visualizing pairwise relationships between numerical variables in a dataset, helping to quickly explore patterns, correlations, and distributions.
#12.What is the purpose of the describe() function in Pandas?
The `describe()` function in Pandas provides summary statistics of a DataFrame’s numeric columns, like count, mean, std, min, and quartiles, giving a quick overview of the dataset’s distribution.
#13.Why is handling missing data important in Pandas?
Handling missing data in Pandas is important to ensure accurate analysis, as missing values can distort results, lead to errors, or reduce the quality of insights if not properly addressed.
#14.What are the benefits of using Plotly for data visualization?
Plotly offers interactive, high-quality visualizations with features like zoom, hover, and export options. It’s great for web integration, easy to use with Pandas, and supports a wide range of chart types.
#15.How does NumPy handle multidimensional arrays?
NumPy handles multidimensional arrays using the `ndarray` object, which allows efficient storage and operations on arrays of any dimension, enabling complex numerical computations across multiple axes.
#16.What is the role of Bokeh in data visualization?
Bokeh is a Python library for creating interactive, web-based visualizations. It’s designed to handle large datasets efficiently and allows dynamic updates, making it ideal for dashboards and real-time data apps.
#17.Explain the difference between apply() and map() in Pandas?
In Pandas, `map()` is used for element-wise operations on a Series, while `apply()` can be used on both Series and DataFrames to apply functions along an axis, offering more flexibility for complex operations.
#18.What are some advanced features of NumPy?
Advanced features of NumPy include broadcasting, vectorized operations, masked arrays, advanced indexing, linear algebra functions, random sampling, and integration with C/C++ for high performance.
#19.How does Pandas simplify time series analysis?
Pandas simplifies time series analysis with built-in date/time types, resampling, frequency conversion, rolling window calculations, and powerful indexing for easy handling of temporal data.
#20.What is the role of a pivot table in Pandas?
A pivot table in Pandas summarizes data by grouping and aggregating values across categories, making it easier to analyze patterns and compare different subsets of data.
#21.Why is NumPy’s array slicing faster than Python’s list slicing?
NumPy’s array slicing is faster than Python’s list slicing because NumPy uses contiguous memory and optimized C-based operations, avoiding the overhead of Python’s dynamic typing and object handling.
#22.What are some common use cases for Seaborn?
Common use cases for Seaborn include visualizing distributions (histograms, KDEs), relationships (scatter, line plots), categorical data (box, bar plots), and statistical summaries (heatmaps, pairplots) with built-in themes and color palettes.

```python
#1.How do you create a 2D NumPy array and calculate the sum of each row?
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sums = np.sum(arr, axis=1)
print(row_sums)
```

```python
#2.Write a Pandas script to find the mean of a specific column in a DataFrame?
import pandas as pd

df = pd.DataFrame({'A': [10, 20, 30], 'B': [5, 15, 25]})
mean_A = df['A'].mean()
print(mean_A)

```

```python
#3.Create a scatter plot using Matplotlib
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
#4.A How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap)
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [4, 3, 2, 1],
    'C': [2, 3, 4, 5]
})

# Correlation matrix
corr = df.corr()

# Heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
#5.Generate a bar plot using Plotly?
import plotly.express as px

data = {'Category': ['A', 'B', 'C'], 'Values': [10, 20, 15]}
fig = px.bar(data, x='Category', y='Values', title='Bar Plot')
fig.show()
```

```python
#8.Create a line plot with multiple lines using Matplotlib
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [5, 15, 20, 25]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Line Plot')
plt.legend()
plt.show()
```

```python
#10.Create a histogram using Seaborn to visualize a distribution.
import seaborn as sns
import matplotlib.pyplot as plt

data = [10, 20, 20, 30, 30, 30, 40, 50]
sns.histplot(data, bins=5, kde=True)
plt.title('Histogram with KDE')
plt.show()
```
